```python
# Highly anonymous data

# "ind" "car" "reg" "calc" mean something
ind_cols = [col for col in train_df.columns if 'ind' in col]
car_cols = [col for col in train_df.columns if 'car' in col]
reg_cols = [col for col in train_df.columns if 'reg' in col]
calc_cols = [col for col in train_df.columns if 'calc' in col]

# "bin" means binary data, and "cat" means categorical data
bin_cols = [col for col in train_df.columns if 'bin' in col]
cat_cols = [col for col in train_df.columns if 'cat' in col]

numeric_cols =  [col for col in train_df.columns if ('cat' not in col and 'bin' not in col)]
```
